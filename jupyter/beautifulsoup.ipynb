```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.geeksforgeeks.org/numpy/python-numpy/"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
texts1 = soup.find_all("h1")
texts2 = soup.find_all("h2")
texts3 = soup.find_all("h3")
```

```python
print("#### Header 1 ###########")
for line in texts1:
    print(line.text.strip())
print("#### Header 2 ###########")
for line in texts2:
    print(line.text.strip())
print("#### Header 3 ###########")
for line in texts3:
    print(line.text.strip())
```
