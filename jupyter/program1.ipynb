```python
l1=[]
tables=[]
for i in range(1,21):
    l1.append(i)
for i in range(1,6):
    df = pd.DataFrame({
        f"{i}":l1,
        f"x{l1[i]}":[i]*20,
        f"={i}": [x*i for x in l1]
    })
    tables.append(df)


space = pd.DataFrame([["                 "]]*20)

final = []
for table in tables:
    final.append(table)
    final.append(space)

final.pop()

final = pd.concat(final,axis=1)


# print(final)

print(final.to_string(index=False,header=False))
```

```python
import pandas as pd

l1 = list(range(1, 21))
tables = []

for i in range(1, 6):
    df = pd.DataFrame({
        f"{i}": l1,
        f"x{i}": [i] * 20,
        f"={i}": [x * i for x in l1]
    })
    tables.append(df)

# blank spacer (2 empty columns)
space = pd.DataFrame([["", ""]] * 20)

# insert space between tables
final_tables = []
for t in tables:
    final_tables.append(t)
    final_tables.append(space)
# print(final_tables)
# remove last extra space
final_tables.pop()

final_df = pd.concat(final_tables, axis=1)

print(final_df.to_string(index=False, header=False))
```
